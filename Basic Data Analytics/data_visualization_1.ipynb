## 1\. Preço do diamante por tipo de corte

```
import seaborn as sns

data = sns.load_dataset("diamonds")
data.head()
```

```
# gráfico do exercício 1

diamonds = sns.barplot(data=data,x='price',y='cut',palette='pastel')
diamonds.set(title='Preço do Diamante Por Tipo de Corte', xlabel='Preço Médio(USD)', ylabel="Tipo de Corte")
diamonds
```

**Insight do gráfico 1**:

O preço médio de cada corte do diamante, logicamente é distinto. O maior preço médio, é o do corte Premium, pela lógica, superando 4.000 dólares.

O corte Premium é o mais sofisticado, logo, o mais caro.

Já o corte Ideal é o mais barato, por ser um corte mais simples, e mais comum.

Porém, entre o corte Ideal e o Premium, existem outros cortes que são importantes, como o Good, o Very Good, e o Fair.

O Fair fica no segundo colocado, superando também a casa dos 4.000 dólares, assim como o Premium.

O corte Very Good, no terceiro colocado, chega perto dos 4.000 dólares mas não chega a superá-lo.

O Good, com o preço médio bem próximo ao Very Good, ocupa a quarta colocação. Também chega muito próximo aos 4.000 dólares.

## 2\. Número de passageiros em dezembro por ano

```
import seaborn as sns

data = sns.load_dataset("flights")
data.head()
```

```
# gráfico do exercício 2

flights_december = data[data['month']=='Dec']
flights_december_grafic = sns.lineplot(data=flights_december, x='year',y='passengers')

flights_december_grafic.set(title="Números de Passageiros em Dezembro Por Ano",xlabel='Ano',ylabel='Número de Passageiros')
```

**Insight do gráfico 2**: Usando um gráfico de linhas para visualizar as tendências ao longo dos anos, observa-se que, ao passar dos anos, o número de passageiros em dezembro aumenta em relação ao mês de dezembro do ano anterior, deixando claro que, em dezembro de cada ano, o números de pessoas que compram uma passagem aérea vão continuar aumentando em comparação ao ano antecedente.

## 3\. Numero de passageiros por mês entre 1949 e 1959

```
import seaborn as sns

data = sns.load_dataset("flights")
data.head()
```

```
# gráfico do exercício 3

passengers_per_month_between_1949_1959 = data.query('year >= 1949 and year <=1959')
passengers_per_month_between_1949_1959_grafic = sns.lineplot(data=passengers_per_month_between_1949_1959,x='month',y='passengers',hue = 'year',palette='Set1')
passengers_per_month_between_1949_1959_grafic.set(title='Numero de passageiros por mês entre 1949 e 1959',xlabel='Mês',ylabel='Número de Passageiros')
passengers_per_month_between_1949_1959_grafic.get_legend().set_title("Ano")
```

**Insight do gráfico 3**: Usando o gráfico de linhas para visualizar a tendência do número de passageiros por mês a cada ano, é nítido que ao passar dos anos este número cresce. Logo, a tendência é que, nos meses dos próximos anos, um número maior de pessoas em relação aos anos antecedentes irão comprar passagens aéreas.

Porém, os meses que mais tem compras de passagens aéreas, são os meses de Julho e Agosto.

Paralelamente, os meses que menos tem compras de passagens aéreas, são os meses de Janeiro, Fevereiro, e Novembro.
